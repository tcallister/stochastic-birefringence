```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('../plotting.mplstyle')
import h5py
from matplotlib import cm
import numpy as np

blue = "#1f78b4"
green = "#33a02c"
red = "#e31a1c"
```

```python
# Load outputs from individual baseline runs

def read_output(filepath):
    
    # Read hdf file
    hfile = h5py.File(filepath,'r')
    
    # Get joint 2D posteriors
    resultsDict = {k:v[()] for k,v in hfile['result'].items()}
    
    return resultsDict

HLO1 = read_output('./../data/fixed_rate_delayedSFR_HLO1.hdf')
HLO2 = read_output('./../data/fixed_rate_delayedSFR_HLO2.hdf')
HLO3 = read_output('./../data/fixed_rate_delayedSFR_HLO3.hdf')
HVO3 = read_output('./../data/fixed_rate_delayedSFR_HVO3.hdf')
LVO3 = read_output('./../data/fixed_rate_delayedSFR_LVO3.hdf')
full = read_output('./../data/fixed_rate_delayedSFR.hdf')
```

```python
# For purposes of normalizing colormaps, get the global max probability from all analyses
pmax = np.max(np.concatenate([HLO1['probabilities'],
                              HLO2['probabilities'],
                              HLO3['probabilities'],
                              HVO3['probabilities'],
                              LVO3['probabilities']]))
```

```python
fig = plt.figure(figsize=(8,2.25))

ax = fig.add_subplot(151)
ax.set_rasterization_zorder(2)
cmap = cm.Blues
cmap.set_under("white")
c = ax.pcolormesh(HLO1['kappa_dcs_2D'],HLO1['kappa_zs_2D'],HLO1['probabilities'].T,
                  cmap=cmap,vmin=1e-3,zorder=1,vmax=pmax)
ax.set_xlabel('$\kappa_D$ [$\mathrm{Gpc}^{-1}$]',fontsize=13)
ax.set_ylabel('$\kappa_z$')
ax.text(0.5, 0.86, 'HL O1',horizontalalignment='center',transform=ax.transAxes,fontsize=12,color="black")
ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.8,0.8)

ax = fig.add_subplot(152)
ax.set_rasterization_zorder(2)
cmap = cm.Blues
cmap.set_under("white")
c = ax.pcolormesh(HLO2['kappa_dcs_2D'],HLO2['kappa_zs_2D'],HLO2['probabilities'].T,cmap=cmap,vmin=1e-6,zorder=1,vmax=pmax)
ax.set_xlabel('$\kappa_D$ [$\mathrm{Gpc}^{-1}$]',fontsize=13)
ax.text(0.5, 0.86, 'HL O2',horizontalalignment='center',transform=ax.transAxes,fontsize=12,color="black")
ax.set_yticklabels([])
ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.8,0.8)

ax = fig.add_subplot(153)
ax.set_rasterization_zorder(2)
cmap = cm.Blues
cmap.set_under("white")
c = ax.pcolormesh(HLO3['kappa_dcs_2D'],HLO3['kappa_zs_2D'],HLO3['probabilities'].T,cmap=cmap,vmin=1e-6,zorder=1,vmax=pmax)
ax.set_xlabel('$\kappa_D$ [$\mathrm{Gpc}^{-1}$]',fontsize=13)
ax.text(0.5, 0.86, 'HL O3',horizontalalignment='center',transform=ax.transAxes,fontsize=12,color="black")
ax.set_yticklabels([])
ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.8,0.8)

ax = fig.add_subplot(154)
ax.set_rasterization_zorder(2)
cmap = cm.Reds
cmap.set_under("white")
c = ax.pcolormesh(HVO3['kappa_dcs_2D'],HVO3['kappa_zs_2D'],HVO3['probabilities'].T,cmap=cmap,vmin=1e-6,zorder=1,vmax=pmax)
ax.set_xlabel('$\kappa_D$ [$\mathrm{Gpc}^{-1}$]',fontsize=13)
ax.text(0.5, 0.86, 'HV O3',horizontalalignment='center',transform=ax.transAxes,fontsize=12,color="black")
ax.set_yticklabels([])
ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.8,0.8)

ax = fig.add_subplot(155)
ax.set_rasterization_zorder(2)
cmap = cm.Greens
cmap.set_under("white")
c = ax.pcolormesh(LVO3['kappa_dcs_2D'],LVO3['kappa_zs_2D'],LVO3['probabilities'].T,cmap=cmap,vmin=1e-6,zorder=1,vmax=pmax)
ax.set_xlabel('$\kappa_D$ [$\mathrm{Gpc}^{-1}$]',fontsize=13)
ax.text(0.5, 0.86, 'LV O3',horizontalalignment='center',transform=ax.transAxes,fontsize=12,color="black")
ax.set_yticklabels([])
ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.8,0.8)

plt.tight_layout()
plt.subplots_adjust(wspace=0.25)
plt.savefig('figure_11.pdf',bbox_inches='tight',dpi=200)
plt.show()
```

```python
def get_max_likelihood_params(resultDict):
    
    dc_argmax,z_argmax = np.unravel_index(np.argmax(resultDict['probabilities']),resultDict['probabilities'].shape)
    kappa_dc_max = resultDict['kappa_dcs_2D'][dc_argmax]
    kappa_z_max = resultDict['kappa_zs_2D'][z_argmax]
    
    return kappa_dc_max,kappa_z_max
```

```python
get_max_likelihood_params(HLO3),get_max_likelihood_params(HVO3),get_max_likelihood_params(full)
```
